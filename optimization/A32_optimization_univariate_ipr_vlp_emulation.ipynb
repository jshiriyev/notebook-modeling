```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
from scipy import optimize
```

```python
def circle(x,radius):
  return np.sqrt(radius**2-x**2)
```

```python
def line(x,slope,intercept):
  return slope*x+intercept
```

```python
def objective(x,radius,slope,intercept):
  yc = circle(x,radius)
  yl = line(x,slope,intercept)
  return (yc-yl)**2
```

```python
sol = optimize.minimize(objective,0,args=(5,2,2),bounds=optimize.Bounds(0,5))
```

```python
x = np.linspace(0,5,1000)
```

```python
plt.plot(x,circle(x,5),label="circle")
plt.plot(x,line(x,2,2),label='line')

plt.scatter(sol.x,circle(sol.x,5),label="Intersection")

plt.xlabel("x-values")
plt.ylabel("y-values")

plt.ylim(ymax=20)

plt.legend()

plt.show()
```
