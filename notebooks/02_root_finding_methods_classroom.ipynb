```python
from scipy.optimize import root_scalar
```

```python
def f(x):
  return x**3+x-1

def g(x):
  return (1-x)**(1/3)
```

```python
root_scalar(f,method='bisect',bracket=[0,2])
```

```python
Niter = 100

xn = 0.5

epsilon = 1e-6

for i in range(Niter):
  xn = g(xn)

  if abs(f(xn))<epsilon:
    break

  print(i,xn,f(xn))
```

```python
def f(x):
    return x**3-4*x-9  # The function

def df(x):
    return 3*x**2 - 4  # The derivative
```

```python
def newton(function,derivative,x0,epsilon,max_iter):
  xn = x0
  for n in range(max_iter):
    fxn = function(xn)
    print(n,xn,fxn)
    if abs(fxn) < epsilon:
      print('Found solution after',n,'iterations!')
      return xn
    xn -= fxn/derivative(xn)
  else:
    print('Exceeded maximum iterations. No solution found.')
```

```python
newton(f,df,0.1,1e-6,100)
```
