```python
def relative_error(true_value,approx_value):
  return abs(true_value-approx_value)/abs(true_value)
```

```python
from math import factorial
```

```python
def exponential(x,n_term=50):

  sum_exp = 0.

  x = abs(x)

  for n in range(n_term):
    sum_exp += (x**n)/factorial(n)

  return 1/sum_exp
```

```python
x_values = np.linspace(-20,0,100)
```

```python
true_values = np.exp(x_values)
appr_values = np.array([exponential(x) for x in x_values])
```

```python
errors = relative_error(true_values,appr_values)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.semilogy(x_values,errors)
plt.show()
```

```python
def quadratic(x,a,b,c):
  return a*x**2+b*x+c
```

```python
def naive_solver(a,b,c):

  D = b**2-4*a*c

  if D < 0:
    return None
  elif D == 0:
    x1,x2 = -b/(2*a),-b/(2*a)
  else:
    x1 = (-b+np.sqrt(D))/(2*a)
    x2 = (-b-np.sqrt(D))/(2*a)

  return x1, x2
```

```python
def stable_solver(a,b,c):

  D = b**2-4*a*c

  if D < 0:
    return None
  elif D == 0:
    x1,x2 = -b/(2*a),-b/(2*a)
  elif b>=0:
    x1 = 2*c/(-b-np.sqrt(D))
    x2 = (-b-np.sqrt(D))/(2*a)
  else:
    x1 = (-b+np.sqrt(D))/(2*a)
    x2 = (2*c)/(-b+np.sqrt(D))

  return x1, x2
```

```python
print(naive_solver(1,200,1e-14))
print(stable_solver(1,200,1e-14))

print(quadratic(0.,1,200,1e-14))
print(quadratic(-5e-17,1,200,1e-14))
```
